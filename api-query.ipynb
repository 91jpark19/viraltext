```python
import requests
import time
from datetime import timedelta
from datetime import datetime
import pandas as pd
import urllib.request
from tqdm import tqdm
import json
import asyncio
```

```python
def api_query(arg1, arg2, page_no):
    lccn=[]
    text=[]
    s=requests.Session()
    userName={'userName':'JP'}
    location={'location':'Chicago'}
    r=s.get('https://chroniclingamerica.loc.gov/search/pages/results/?andtext='+arg1+'&ortext='+arg2+'&page='+str(page_no)+'&format=json')
    # time.sleep(1)
    for item in range(0,len(r.json()['items'])):
        lccn.append(r.json()['items'][item]['id'])
        text.append(r.json()['items'][item]['ocr_eng'])
    return pd.DataFrame({'lccn': lccn, 'text': text})
```

```python
book=[]
pause=0
exclude_page=[]
for i in tqdm(range(0,22545)):
    if i==0:
        pass
    else:
        try:
            pause+=1
            page_result=api_query('mob', 'attack', i)
            book.append(page_result)
            if pause%50==0:
                time.sleep(5)
        except:
            time.sleep(3)
            exclude_page.append(i)        
```
